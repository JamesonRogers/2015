```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import pandas as pd

# PART A: GENERATE THE PRIORS
def prior(theta1, theta2):
    a=norm.pdf(theta1, loc= 31, scale=30)
    b=norm.pdf(theta2, loc=-0.5, scale=1)
    c=a*b
    return c

# PART B: CALCULATE THE POSTERIOR
data=np.loadtxt('chall.txt')
T=data[:,0]
p_fail=data[:,1]

print T
print p_fail

def mylikelihood(pfail, temp, thet1, thet2):
    logistic_fun= 1./(1 + np.exp(-1*(thet1 + thet2*temp)))
    p=np.prod(logistic_fun**pfail*(1-logistic_fun)**(1-pfail))
    return p


# Now lets use rejection to draw random samples, t1 and t2 from the posterior

foo=200
x = np.linspace(0,50,foo)
y = np.linspace(-1,0.3,foo)
xx,yy = np.meshgrid(x,y)
zz=np.zeros((foo,foo))
for i in np.arange(0,foo):
    for j in np.arange(0,foo):
        zz[i,j] = mylikelihood(p_fail, T, xx[i,j], yy[i,j])

plt.contourf(xx,yy,zz)
plt.xlabel('t_1')
plt.ylabel('t_2')
plt.show()

print np.max(zz)
```

Now, we are given an asymmetric function (a Gaussian shifged from the origin) as a proposal distribution. The success of the algorithm depends on the random walk being symmetric. So in general we want that every time we jump to a new position, the probability of going to this new position from our current position be the same of the reverse transition. In other words, we want $q(x|x') = q(x'|x)$. In the case of a Gaussian distribution, this happens naturally, and so we only need to use the ratio of propabilities $P(x')/P(x)$ in the acceptance ratio. In the case of asymmetric proposal functions, however, we need to include the ratio $q(x'|x)/q(x|x')$ in the acceptance ratio. For comparison and illustrative purposes, let's first use a Gaussian proposal.

```python
# First symmetric proposal (Metropolis)

x0=[5.,-5.]     # Starting point
cnt=0           # Initialize counts
Ns=100000       # Number of samples
xsample=np.zeros((2,Ns))
xsample[:,0]=x0
gam=2           # Gaussian parameter for the proposal distribution
acceptcnt=0
```

```python
lprop_old = np.log(func(xsample[0,0],xsample[1,0]))

# Here we implement the Metropolis algorithm. The proposal distribution is Gaussian, and so we use.
# numpy.random.normal

while cnt+1 < Ns:
    xstar = xsample[:,cnt]+np.random.normal(0,gam,2)                # The next step in the chain is sampled randomly from a Gaussian
    lprop_new = np.log(func(xstar[0],xstar[1]))
    
    if (lprop_new - lprop_old) > np.log(np.random.uniform(0,1,1)):  # Acceptance ratio. Notice that here only the ratio of the evaluated functions matters.
        x0=xstar
        lprop_old=lprop_new
        acceptcnt += 1
    cnt += 1
    xsample[:,cnt]=x0
```

In general, the burn-in factor is necessary because the initial samples might follow a very different distribution as the target distribution. We therefore want to get rid of these initial samples. How large this factor is depends on where in the parameter space we start sampling. In this, however, our proposal is well matched to the target distribution, and we do not start the Markov Chain too far from the origin, so the burn-in factor is not so important.

Now we turn to the case where the proposal function is an asymmetric function. In this case, the algorithm must include the Hastings factor (tha ratio of the transition probabilities) to correct for the asymmetry of the distribution. Otherwise, the detailed balance property of the Markov chain is not fulfilled, and we can end up oversampling a aprticular region of the parameter space. In other words, if we do not correct for the asymmetry of the proposal distribution, the sampling of the target distribution will be skewed towards a preferential direction.

```python
# Now asymmetric (Metropolis-Hastings)

x0=[5,-5.]    # Initial location
cnt=0         # Initialize counts
Ns=100000     # Number of samples
xsample=np.zeros((2,Ns))
xsample[:,0]=x0
beta = 1.0
gam1 = 1.0*beta
gam2 = 1.5*beta
acceptcnt=0
```

```python
# Here we define the proposal function explicitely, in order to calculate the transition ratios.

g = lambda x,y,gam1,gam2: (1/np.sqrt(2*np.pi*gam1*gam2))*np.exp(-(((x-0.1)**2)/(2*gam1**2)+ ((y-0.1)**2)/(2*gam2**2)))

# And now we plot the proposed transition probability

foo=200
x = np.linspace(-5,5,foo)
y = np.linspace(-5,5,foo)
xx,yy = np.meshgrid(x,y) # Define a mesh grid in the region of interest
zz=np.zeros((foo,foo))
for i in np.arange(0,foo):
    for j in np.arange(0,foo):
        zz[i,j] = g(xx[i,j],yy[i,j],gam1,gam2) # Populate the mesh

levels = np.arange(0,0.35,0.05)        
plt.contour(xx,yy,zz,levels)
axes().set_aspect('equal', 'datalim')
```

```python
lprop_old = np.log(func(xsample[0,0],xsample[1,0]))  # Function evaluated at the starting position

incs = []

# Markov Chain:
while cnt+1 < Ns:
    increment = np.array([np.mean(np.random.normal(0.1,gam1,1)),np.mean(np.random.normal(0.1,gam2,1))]) # The proposed step, sampled from the transition probability
    xstar = xsample[:,cnt]+increment   # New proposed position
        
    lprop_new = np.log(func(xstar[0],xstar[1]))  # Function evaluated at the new position
    
    # Now we do the Metropolis-Hastings
    # Notice that in this case we need to include an additional factor in the acceptance ratio. This
    # is to compensate for the asymmetry of the proposal function.
    
    # Now we evaluate the Metropolis-Hastings acceptance
    if min(np.log(1),lprop_new-lprop_old+np.log(g(-increment[0],-increment[1],gam1,gam2))-np.log(g(increment[0],increment[1],gam1,gam2))) >= np.log(np.random.uniform(0,1,1)):
        x0=xstar
        lprop_old=lprop_new
        acceptcnt += 1
    else:
        x0=xsample[:,cnt]
        
    cnt += 1
    xsample[:,cnt]=x0
    

print 'The acceptance rate is: ',acceptcnt/float(Ns)
```

```python
# Let us plot the result (histogram of our samples)

xout=xsample[:,Ns/10.:] # Just for the sake of it, let's cut 10% of the sample for Burn-in 
plt.hist2d(xout[0,:], xout[1,:], normed=True, bins=100)
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.show()
```

For the same number of steps, we get a larger acceptance rate than in the symmetric case. We are able to sample the four individual flux peaks, which clearly shows that the algorithm is not skewed (it does not get stuck in local maxima)

### Question 3: Integrating nasty functions with MC

(a) What should be the properties of $g(x)$ and $h(x)$ for our MC approximation to be as accurate as possible? Can our integrand be split as a product of such functions? What part would you use as $g(x)$ and why?

The new integrand $h(x)$ should be "flatter" than, i.e., its variance should be reduced w.r.t., the original integrand, in order to improve the way in which our random sampling reproduces steep changes in the function as we evaluate it. Also, when possible, the distribution $g(x)$ from which we will draw our samples should be easy to normalize, in order to guarantee that the integral in question is not scaled by an undesired factor. In our particular case, after inspecting the function we easily conclude that a good candidate for a sample distribution is $1/[2+(x-4)^2]$, since we can calculate its integral analytically and then invert it to obtain the normalization factor, which happens to be:

$C = \left \{\frac{1}{\sqrt{2}}\left[\frac{\pi}{2}-\tan ^{-1}\left(\frac{-4}{\sqrt{2}}\right)\right]\right \}^{-1}$

We can invert $g(x)$ analytically to sample the $x_i$ from the distribution g$*$, where $ g* = Cg $

(b) Use $1/(2+(x-4)^2)$ as your $g(x)$ (remember that you need to normalize it). Implement a Metropolis-Hastings algorithm that has as a stationary distribution to sample from, $g(x)$.  Run the simulation 50 times for 150,000 points. Report the value of the integral $\hat{I}$ and the VAR(${\hat{I}}$).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# We first define the functions.

g_norm = ((1./np.sqrt(2))*(np.pi/2-np.arctan(-4./(np.sqrt(2)))))**(-1)
g = lambda x: g_norm/(2+(x-4)**2)
h = lambda x: (x**4)*np.sin(np.sqrt(np.log(x+1)))*np.exp(-x)
f = lambda x: g(x)*h(x)


x = np.arange(0,15, 0.05)

# Let us first plot the functions to see how they look like.

plot(x,f(x),color='blue')
plot(x,g(x),color='red')
plot(x,h(x),color='green')
```

```python
gstar = lambda x: g_norm/(2+(x-4)**2) # target stationary distribution, appropriately normalized
propfun = lambda x: x + np.random.uniform(-1,1) # the proposal adds a realization drawn from U(-1,1) to the current point x

#Let's first use 15,000 points.

N = 1.5*10**5 # number of points
Ns = 50 # number of runs

IMH = np.zeros(Ns) # approximation for I, one for each run
```

```python
for i in np.arange(0,Ns):
    xvector = np.zeros(N) # for storing final settled on values at each step
    xprev = np.random.uniform(0,1) # random initial condition, may as well draw from U[0,1]. 
    unifvector = np.random.uniform(0,1,N) # vector to decide if we accept or reject proposed point

    for j in np.arange(0,N):
        xstar = propfun(xprev) #propose new value based on old one
        if xstar >= 0:
            if unifvector[j] <= min(1,gstar(xstar)/gstar(xprev)):
                xprev = xstar # move to new value
                xvector[j] = xprev #move to new value and store it
            else:
                xvector[j] = xprev # remain at old value and store it
        else:
            xvector[j] = xprev # remain at old value and store it
            
    xvector = xvector[0.2*N:] # burn first 20% of simulation
    IMH[i] = np.mean((1/g_norm)*(xvector**4)*np.sin(np.sqrt(np.log(xvector+1)))*np.exp(-xvector)) # approximation for I
    
print np.mean(IMH),np.var(IMH)
plt.hist(IMH)
```

```python
# With fewer points

N = 1.5*10**4 # number of points

for i in np.arange(0,Ns):
    xvector = np.zeros(N) # for storing final settled on values at each step
    xprev = np.random.uniform(0,1) # random initial condition, may as well draw from U[0,1]. 
    unifvector = np.random.uniform(0,1,N) # vector to decide if we accept or reject proposed point

    for j in np.arange(0,N):
        xstar = propfun(xprev) #propose new value based on old one
        if xstar >= 0:
            if unifvector[j] <= min(1,gstar(xstar)/gstar(xprev)):
                xprev = xstar # move to new value
                xvector[j] = xprev #move to new value and store it
            else:
                xvector[j] = xprev # remain at old value and store it
        else:
            xvector[j] = xprev # remain at old value and store it
            
    xvector = xvector[0.2*N:] # burn first 20% of simulation
    IMH[i] = np.mean((1/g_norm)*(xvector**4)*np.sin(np.sqrt(np.log(xvector+1)))*np.exp(-xvector)) # approximation for I
    
print np.mean(IMH),np.var(IMH)
plt.hist(IMH)
```

We can see that the variance is reduced when we use more points. As we have seen in class, the variance of a Monte Carlo estimate depends on the number of points N we sample and on the variance of our function $f$: $\sigma_{\bf{MC}}^2=\sigma_{\bf{f}}/N$. According to this, our ten-fold increase in the number of samples should correspond to about a factor of 3 decrease in the variance. With 50 repetitions of the experiment the numbers should not exactly match this prediction, but they should be close.

The normalization of the gamma function $\Gamma(2) = xe^{-x}$ is 1. We can therefore approximate the integral as:

$\int_{0}^{\infty} \frac{x^4\, \sin\left(\sqrt{\ln{(x+1)}}\right)e^{-x}}{2+(x-4)^2} \, dx = \frac{1}{N} \sum\limits_{i=1}^{N} \frac{x^3\, \sin\left(\sqrt{\ln{(x+1)}}\right)}{2+(x-4)^2} $

where the $x_i$ are sampled from the $\Gamma(2)$ function. Let us implement this numerically.

```python
Ns = 15000

IIS = np.zeros(Ns) # approximation for I, one for each run

func = lambda x: (x**3) * (np.sin(np.sqrt(np.log(x+1))))/(2+(x-4)**2) # Original f(x)/G(2,1)pdf

for i in np.arange(0,Ns):
    xis=np.random.gamma(2, 1, N)
    IIS[i] = np.mean(func(xis))
    
print np.mean(IIS), np.var(IIS)
plt.hist(IIS)
```

The variance is significantly reduced when we use $xe^{-x}$ as a distribution to draw random variate from. As you remember from class, the square of the standard deviation associated with a MC integration is proportional to the variance of the function being evaluated and inversely proportional to the number of points we sample. In this case, we have used the same numbers of points as before, but by choosing $xe^{-x}$ as a distribution to draw random variate from, the new integrand is even flatter than in the case when we use $g(x) = 1/[2+(x-4)^2]$. This is the reason why the variance is reduced.
