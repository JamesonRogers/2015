### Problem 1

The inertia of the gift is given by $I=\int dx\int dy \int dz\rho(x,y,z) r^2(x,y,z)$. Because it's difficult to sample uniform distributed points in the shape, we will sample uniform distributed points from the cube(or box) that contains the gift. If the point is not inside the gift, we will set $\rho(x,y,z)=0$. A more detailed description is as follows:

The point are generated by:
$$x\sim U(-\sqrt{2},\sqrt{2})$$
$$y\sim U(-\sqrt{2},\sqrt{2})$$
$$z\sim U(-1,1)$$
The square of the perpendicular distance to the z-axis is:
$$r^2(x,y,z) = x^2+y^2$$
The density is given by:
$$\rho(x,y,z)=\rho_1 \mbox{ if } x^2+y^2+z^2\leq r_1^2$$
$$\rho(x,y,z)=\rho_2 \mbox{ if }x^2+y^2+z^2\geq r_1^2\mbox{ and }x^2+y^2\leq r^2(z)$$
$$\rho(x,y,z)=0 \mbox{ otherwise}$$

For each simulation, we gererate $N_{mc}$ points as above, calculate the mean of $\rho(x,y,z)r^2(x,y,z)$ and the times the result by the volume. The volumn is 16 in this problem. The formula for the generic error scaling of Monte Carlo is $\hat{\sigma}_{M}\approx\frac{\sigma}{\sqrt{N}}\propto \frac{1}{\sqrt{N}}$ and we would like to choose $N_{mc}$ such that the estimated accuracy $\frac{\hat{\sigma}_{M}}{\hat{I}}$ be less or equal to $0.1\%$. First, let us try with $N_{mc}=10^5$ and get a rough estimate of $\sigma$.

```python
# density of the gift
rho0, rho1 , rho2 = 0.0, 4.0, 6.0
# radius of the sphere
r1 = 1.0
# size of the cub
x_min, x_max, y_min, y_max, z_min, z_max = -2.0**0.5, 2.0**0.5, -2.0**0.5, 2.0**0.5, -1.0, 1.0
# volumn of the cube
volume = (x_max - x_min)*(y_max - y_min)*(z_max - z_min)
# number of repetition
N_s = 1000
```

```python
# get r(z) give z
def rz(z):
    return np.sqrt(1+abs(z))

# the density function of the cube
def density(x, y, z):
    r2 = rz(z)
    d = np.repeat(rho2, len(x))
    d[x**2+y**2>r2**2] = rho0
    d[x**2+y**2+z**2<r1**2] = rho1
    return d

# get r(x, y, z) given the point
def distance(x, y, z):
    return np.sqrt(x**2+y**2)

# compute inertia if sample N_mc points for MCMC
def Inertia(N_mc):
    x = np.random.random(N_mc)*(x_max-x_min)+x_min
    y = np.random.random(N_mc)*(y_max-y_min)+y_min
    z = np.random.random(N_mc)*(z_max-z_min)+z_min
    
    integrands = (distance(x, y, z)**2*density(x, y, z)*volume).mean()
    return integrands

# compute accuracy for different N_mc
def repeat_mcmc(N_s, N_mc):
    startTime = time.time()
    calIntegrand = np.empty(N_s)
    for i in xrange(N_s):
        calIntegrand [i] = Inertia(N_mc)
        if i% 100 == 0:
            print "On %dth simulation" %i

    aveIntegrand = round(calIntegrand.mean(), 3)
    errIntegrand = round(calIntegrand.std(),3)
    accuracy = round(errIntegrand/aveIntegrand*100, 3)
    timeElapsed = round(time.time() -startTime, 3)
    
    print "I = %s+/- %s" %(aveIntegrand, errIntegrand)
    print "Time elapsed is %s seconds" %timeElapsed
    print "The accuracy is %s%%" %accuracy
```

```python
repeat_mcmc(N_s, 10**5)
```

With $N_{mc}=10^5$, the accuracy is $0.4\%$. If we want to reduce the accuracy to $0.1\%$, we could try $N_{mc}=(0.4/0.1)^2*10^5=1.6*10^6$. The result is as follows:

```python
repeat_mcmc(N_s, 16*10**5)
```

With $N_{mc}=1.6*10^6$, we get an accuracy less than $0.1\%$.

### Problem 2

We will solve this problem by Monte Carlo method, that is we first generate pairs of $(a, h, w)$ that satisfies the distributions and then compute the propotion that satisfies the condition. Because age and (height, weight) are independent, we could generate then seperately. Let $N_{mc}=10^5$ and we will repeat it for $N_s=1000$ times.

```python
# parameters for the probability
Lambda = 0.3
Mu = [30, 25]
Sigma = [[50, 10], [10, 100]]

# parameters for mcmc
N_s, N_mc = 1000,  10**5
```

```python
# generate pairs of (a, h, w)
def generate(N_mc):
    a = np.random.exponential(scale = 1.0/Lambda, size = N_mc)
    h, w = np.random.multivariate_normal(Mu,Sigma,N_mc).T
    return a, h, w

# how many of the pairs satisfy the conditions
def filters(a, w, h):
    pick = np.zeros(len(a))
    pick[a<6] += 1
    pick[h>=20] += 1
    pick[h<=35] += 1
    pick[w<=40] += 1
    return len(pick[pick==4])

# calculate the probability
def calProbability(N_mc):
    a, h, w = generate(N_mc)
    pick = filters(a, w, h)
    return float(pick)/N_mc

# repeat the Monte Carlo and compute the accuracy
def repeat_mcmc(N_s, N_mc):
    startTime = time.time()
    calProb= np.empty(N_s)
    for i in xrange(N_s):
        calProb [i] = calProbability(N_mc)
        if i% 100 == 0:
            print "On %dth simulation" %i

    aveProbability = round(calProb.mean(),3)
    errProbability = round(calProb.std(),3)
    timeElapsed = round(time.time() - startTime,3)
    accuracy = round(errProbability/aveProbability*100, 3)

    print "p = %s+/- %s" %(aveProbability, errProbability)
    print "Time elapsed is %s seconds" %timeElapsed
    print "The accuracy is %s%%" %accuracy
```

```python
repeat_mcmc(N_s, N_mc)
```

So the probability Paparus will find a desire puppy is 0.534.

### Problem 3

For rejection sampling, we generate $(x,y)$ from another distribution $F(x,y)$ that satisfies $P(x,y)\leq MF(x,y)$ for all $(x,y)$. We will accept the sample with probability $\frac{P(x,y)}{MF(x,y)}$. In this problem, we have $P(x,y)\propto exp(-x^2-y^2-x^2y^2),-2\leq x\leq 2,-2\leq y\leq 2$. We could first try $F(x,y)\propto 1$, which is a uniform distribution. Then we have $M=1$.
Let $N_+$ be the number of samples accepted by the rejection sampling and $N_-$ be the number of rejection. Our goal is to increase $\frac{N_+}{N_-}$. We will try the sampling for $N_s=1000$ times and each with $N_{mc}=10^5$ samples.

```python
# parameters for Monte Carlo
M = 1
N_s = 1000
N_mc = 10**5
```

```python
# F(x,y)
def F(x,y):
    return 1

# P(x,y)
def P(x,y):
    return np.exp(-x**2-y**2-x**2*y**2)

# accept probability
def prob(x,y):
    return P(x,y)/(F(x,y)*M)

# generate x, y
def generate(N_mc):
    x = np.random.random(N_mc)*4-2
    y = np.random.random(N_mc)*4-2
    return x, y                      

# rejection sampling
def rejection(N_mc):
    x, y =  generate(N_mc)
    random = np.random.random(N_mc)
    acc = prob(x,y)
    x = x[random<acc]
    
    Nplus = len(x)
    Nminus = N_mc - len(x)
    
    return float(Nplus)/Nminus

# repeat the Monte Carlo and compute the accuracy
def repeat_mcmc(N_s, N_mc):
    startTime = time.time()
    calRatio= np.empty(N_s)
    for i in xrange(N_s):
        calRatio [i] = rejection(N_mc)
        if i% 100 == 0:
            print "On %dth simulation" %i

    aveRatio = round(calRatio.mean(),3)
    errRatio = round(calRatio.std(),3)
    timeElapsed = round(time.time() -startTime,3)
    accuracy = round(errRatio/aveRatio*100,3)

    print "r = %s+/- %s" %(aveRatio, errRatio)
    print "Time elapsed is %s seconds" %timeElapsed
    print "The accuracy is %s%%" %accuracy
```

#### Method 1: stratified sampling

First, we could use a stratified rejection sampling, that is we slice the space of $[-2,2]^2$ into smaller squares and in each square, we find $F(x,y)=c$ such that $P(x,y)\leq c$. We cut the space into $4N^2$ pieces and square (i,j) is $[\frac{i-1}{N}-2, \frac{i}{N}-2]\times [\frac{j-1}{N}-2, \frac{j}{N}-2]$, $1\leq i\leq 2N, 1\leq j\leq 2N$.
For $(x,y)$ in square $(i,j)$, let $x_1 = \min(|\frac{i-1}{N}-2|, |\frac{i}{N}-2|)$ and $y_1 = \min(|\frac{j-1}{N}-2|, |\frac{j}{N}-2|)$, then we have $F(x,y)\leq P(x_1,y_1)$ so and we also have $M=1$. Will sample from $F(x,y)$ and then perform the rejection sampling as above.

```python
# parameters for Monte Carlo
M = 1
N1 = 1000
N2 = 10**5
# parameters for small squares, N could be changed
N = 10
size = 2*N
step = 2.0/N
```

```python
# P(x,y)
def P(x, y):
    return np.exp(-x**2-y**2-x**2*y**2)

# small square and F(x,y), c
def piece(size):
    piecex = np.linspace(-2,2, size+1)
    piecey = np.linspace(-2,2, size+1)
    
    probability = np.zeros(size**2)
    indexx = np.zeros(size**2)
    indexy = np.zeros(size**2)
    for i in xrange(size):
        for j in xrange(size):
            minx = min(abs(piecex[i]), abs(piecex[i+1]))
            miny = min(abs(piecey[j]), abs(piecey[j+1]))
            probability[i*size+j] = prob(minx, miny)
            indexx[i*size+j] = i
            indexy[i*size+j] = j
    return probability, piecex, piecey, indexx, indexy

# F(x,y)
def F(x,y,N):
    x_1 = np.array(abs(x)*N/2, dtype=int)
    y_1 = np.array(abs(y)*N/2, dtype=int)
    x_1 = np.array(x_1, dtype=float)/N*2
    y_1 = np.array(y_1, dtype=float)/N*2
    
    return P(x_1, y_1)
    

# generate from F(x,y)
def generate(N_mc):
    probability, piecex, piecey, indexx, indexy = piece(size)
    tempProb = probability/probability.sum()
    
    location = np.random.choice(size**2, size=N2,p=tempProb)
    locx = np.asarray(indexx[location], dtype=int)
    locy = np.asarray(indexy[location], dtype=int)
    
    x = np.random.random(N2)*step + piecex[locx]
    y = np.random.random(N2)*step + piecey[locy]
    
    return x, y, location, probability

def rejection(N_mc):
    x, y, location, probability = generate(N_mc)
    random = np.random.random(N_mc)
    rej = prob(x,y)/probability[location]
    
    x = x[random<rej]
    
    Nplus = len(x)
    Nminus = N_mc - len(x)
    
    return float(Nplus)/Nminus

# repeat the Monte Carlo and compute the accuracy
def repeat_mcmc(N_s, N_mc):
    startTime = time.time()
    calRatio= np.empty(N1)
    for i in xrange(N_s):
        calRatio [i] = rejection(N_mc)
        if i% 100 == 0:
            print "On %dth simulation" %i

    aveRatio = round(calRatio.mean(), 3)
    errRatio = round(calRatio.std(), 3)
    timeElapsed = round(time.time() -startTime, 3)
    accuracy = round(errRatio/aveRatio*100,3)

    print "r = %s+/- %s" %(aveRatio, errRatio)
    print "Time elapsed is %s seconds" %timeElapsed
    print "The accuracy is %s%%" %accuracy
```

We will show a plot for $N=10$.

```python
N = 10
grid_x = np.linspace(-2,2,50)
grid_y = np.linspace(-2,2,50)
x, y = np.meshgrid(grid_x, grid_y)
z = P(x,y)
z1 = F(x,y,N)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1, 1, 1, projection='3d')
p=ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)
p=ax.plot_wireframe(x, y, z1, rstride=4, cstride=4)
```

```python
repeat_mcmc(N_s, N_mc)
```

We choose 400 smaller squares this time and the ratio increases a lot.

#### Method 2: normal distribution

We could try $F(x,y)\propto exp(-x^2-y^2), -2\leq x\leq 2,-2\leq y\leq 2$ which is a truncated normal distribution. We have $P(x,y)\leq F(x,y)$ for $-2\leq x\leq 2,-2\leq y\leq 2$. In fact it's difficult to sample a certain number(say exact 1000) data points from a truncated normal distribution, we could extend $P(x,y)$ such that
$$P(x,y)\propto exp(-x^2-y^2-x^2y^2),-2\leq x\leq 2,-2\leq y\leq 2$$
$$P(x,y)=0\mbox{ otherwise}$$
Also, we have
$$F(x,y)\propto exp(-x^2-y^2) \mbox{ for all x, y}$$
and we still have 
$$P(x,y)\leq F(x,y)$$
Will sample from $F(x,y)$ and the perform the rejection sampling as above.

```python
# F(x,y)
def F(x,y):
    return np.exp(-x**2-y**2)

# P(x,y)
def P(x,y):
    return np.exp(-x**2-y**2-x**2*y**2)

# accept probability
def prob(x,y):
    return P(x,y)/(F(x,y)*M)

# generate x, y
def generate(N_mc):
    x = np.random.normal(0, np.sqrt(0.5), N_mc)
    y = np.random.normal(0, np.sqrt(0.5), N_mc)
    return x, y                      

# rejection sampling
def rejection(N_mc):
    x, y =  generate(N_mc)
    random = np.random.random(N_mc)
    acc = prob(x,y)
    x = x[random<acc]
    
    Nplus = len(x)
    Nminus = N_mc - len(x)
    
    return float(Nplus)/Nminus

# repeat the Monte Carlo and compute the accuracy
def repeat_mcmc(N_s, N_mc):
    startTime = time.time()
    calRatio= np.empty(N_s)
    for i in xrange(N_s):
        calRatio [i] = rejection(N_mc)
        if i% 100 == 0:
            print "On %dth simulation" %i

    aveRatio = round(calRatio.mean(),3)
    errRatio = round(calRatio.std(),3)
    timeElapsed = round(time.time() -startTime,3)
    accuracy = round(errRatio/aveRatio*100,3)

    print "r = %s+/- %s" %(aveRatio, errRatio)
    print "Time elapsed is %s seconds" %timeElapsed
    print "The accuracy is %s%%" %accuracy
```

We will show a plot of $P(x,y)$ and $F(x,y)$.
